```python
# example usage of class 'COVID19_measures'
# download data from 'https://github.com/amel-github/covid19-interventionmeasures'
# prepare them for use in python data analysis scripts


from imp import reload
import measureclass as mc; reload(mc);

measures = mc.COVID19_measures( download_data = False,
                                measure_level = 2,
                                only_first_dates = False,
                                extend_measure_names = True)


# override default values with explicit calls to 'GetCountryData'
data_austria_L1 = measures.CountryData('Austria', measure_level = 1)
print(data_austria_L1)
```

```python
# iterate over all countrynames and all dictionaries with default values provided
for countryname,measuredata in measures:
    print(countryname)
    for measurename,datelist in measuredata.items():
        print('   {}: '.format(measurename) + ', '.join(datelist))
```

```python
for measure_level in range(1,4):
    print('Measure L{}'.format(measure_level))
    md = {}
    for c in measures.countrylist:
        cd = measures.CountryData(c, measure_level = measure_level, extend_measure_names = True)
        #print(cd)
        for mn, dates in cd.items():
            if not mn in md.keys():
                md[mn] = 0
            md[mn] += 1
    measurelist = list(md.keys())
    measurelist.sort()
    for mn in measurelist:
        print('  {:3d} {:s}'.format(md[mn],mn))
    print()
```
